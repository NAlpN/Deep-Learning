```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('AirPassengers.csv')
data.head()
```

```python
data.rename(columns={'#Passengers' : 'passengers'}, inplace = True)
```

```python
data = data['passengers']
```

```python
type(data)
```

```python
data = np.array(data).reshape(-1,1)
```

```python
type(data)
```

```python
plt.plot(data)
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
len(data)
```

```python
train = data[0:100,:]
test = data[100:,:]
```

```python
def get_data(data, steps):      
    dataX = []
    dataY = []
    for i in range(len(data)-steps-1):
        a = data[i:(i+steps), 0]
        dataX.append(a)
        dataY.append(data[i+steps, 0])
    return np.array(dataX), np.array(dataY)
```

```python
steps = 2
```

```python
X_train, y_train = get_data(train, steps)
X_test, y_test = get_data(test, steps)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
model = Sequential()
model.add(LSTM(128, input_shape = (1, steps)))
model.add(Dense(64))
model.add(Dense(64))                                  
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = scaler.inverse_transform(y_pred)
y_test = y_test.reshape(-1, 1)
y_test = scaler.inverse_transform(y_test)
```

```python
plt.plot(y_test, label = 'Gerçek Yolcu Sayısı')
plt.plot(y_pred, label = 'Tahmin Edilen Yolcu Sayısı')
plt.ylabel('Aylar')
plt.ylabel('Yolcuların sayısı')
plt.legend()
plt.show()
```
